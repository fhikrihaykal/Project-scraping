```python
import pandas as pd
from bs4 import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
#install browser such as chrome or edge
chromedriver_autoinstaller.install()

#use the installed browser for selenium
browser = webdriver.Chrome()
```

```python
browser.get('https://www.rumah.com/panduan-properti/harga-bahan-bangunan-65434')
```

```python
#make it beautiful html
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find('div', class_='ArticlePageTemplate_articlePageContainer__7LtlA container')
```

```python
item = soup.find_all('div', class_='article-table-row')
```

```python
item
```

```python
harga_berat = soup.find_all('div', class_='col-content')
```

```python
harga_berat
```

```python
data_harga =[]
for index in range(3, len(harga_berat)):
    if index %2 == 0:
        data_harga.append(harga_berat[index])
    
```

```python
data_harga
```

```python
data_harga[0]
```

```python
harga_fix = []

for n in range (0, 42):
    if n % 2 != 0:  # Memeriksa apakah indeks tersebut ganjil
        harga_fix.append(data_harga[n].text)  # Menambahkan karakter ganjil ke data_ganjil
```

```python
harga_fix
```

```python
df = pd.DataFrame(harga_fix)
```

```python
df.head(21)
```

```python
ukuran_fix = []

for a in range (0, 42):
    if a % 2 == 0:  # Memeriksa apakah indeks tersebut ganjil
        ukuran_fix.append(data_harga[a].text)  # Menambahkan karakter ganjil ke data_ganjil
```

```python
ukuran_fix
```

```python
df = pd.DataFrame(ukuran_fix)
```

```python
semen = soup.find_all('div', class_='col-header col-header-with-three-or-more')

data_semen =[]
for s in range(0, 21):
    data_semen.append(semen[s].text)
```

```python
data = {
    'Jenis':data_semen,
    'Harga' : harga_fix,
    'Ukuran': ukuran_fix
}
```
